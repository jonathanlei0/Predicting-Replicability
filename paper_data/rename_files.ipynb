```python
import os
import numpy as np
import pandas as pd
import re
```

```python
def clear_special_char(string):
    string = string.replace(':','')
    string = string.replace('/','')
    string = string.replace('"','')
    string = string.replace("'",'')
    string = string.replace("_",'')
    string = string.replace("-",'')
    return string

def resort_dir_by_title(directory, title_list, title_id_dict):
    directory = directory.lower()
    for i, file_name in enumerate(np.sort(os.listdir(directory))):
        print(i)
        pos = file_name.find(']')
        prefix = file_name[: pos+1].lower()
        real_name = clear_special_char(file_name[pos+1:].lower())
        file_name_hash = real_name[:20]
        if ((not (directory in prefix)) 
            or (real_name[0].isdigit() and len(real_name)<10)): 
            continue

        
        #print(file_name_hash)
        found = False
        for j, title in enumerate(title_list):
            if file_name_hash in clear_special_char(title.lower()):
                found = True
                paper_id = title_id_dict[title]
#                 tmp_file = open(f'{directory}/{file_name}','r')
#                 tmp_file.close()
                os.rename(f'{directory}/{file_name}',
                         f'{directory}/[{directory}]{paper_id}.txt')
                #print('Found!\n')
        if not found: 
            print(f"Error: Cannot find the title for {file_name}!\n\n\n")
```

```python
title_list = []
title_dire = 'title/'
paper_id = 0
for file_name in np.sort(os.listdir(title_dire)):
    if file_name[0]=='[' or file_name[0]=='.': continue
    file = open(title_dire+file_name,'r')
    title = file.read()
    file.close()
    title_list.append(title)
    
    
title_list = np.sort(title_list)
    
title_id_dict = {}
id_title_dict = {}
for i, title in enumerate(title_list):
    #print(title)
    title_id_dict[title] = i
    id_title_dict[i] = title
    file = open(f'title/[title]{i}.txt','w')
    file.write(title)
    file.close()
    
id_title_csv = pd.DataFrame.from_dict(id_title_dict, orient='index', columns=['paper_title'])
id_title_csv.to_csv('title/[helper_file]id_title_map.csv')
 
resort_dir_by_title('abstract', title_list, title_id_dict)
resort_dir_by_title('method', title_list, title_id_dict)
        
```
